## Multiclass Classification

```python
from warnings import filterwarnings
filterwarnings('ignore')
```

### Creating a multiclass data

```python
from sklearn.datasets import make_classification
X, Y = make_classification(n_samples=6000, n_features=30, n_informative=15, n_classes=3, random_state=42)
```

```python
X
```

```python
X.shape
```

```python
Y
```

### Convert X and Y to dataframe

```python
cols=[f"x{i+1}" for i in range (30)]
cols
```

```python
import pandas as pd
X=pd.DataFrame(X,columns=cols)
X
```

```python
Y=pd.DataFrame(Y,columns=['target'])
Y
```

```python
Y.value_counts()
```

### Creating preprocessing pipeline for X

```python
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
```

```python
num_pipe=Pipeline(steps=[('impute',SimpleImputer(strategy='median')),('scaler',StandardScaler())])
```

```python
X_pre=num_pipe.fit_transform(X)
X_pre
```

```python
cols=num_pipe.get_feature_names_out()
cols
```

```python
X_pre=pd.DataFrame(X_pre,columns=cols)
X_pre.head()
```

### Apply train test split

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(X_pre,Y,test_size=0.3,random_state=42)
```

```python
xtrain.head()
```

```python
ytrain.head()
```

```python
xtest.head()
```

```python
ytest.head()
```

```python
ytest.value_counts()
```

```python
xtrain.shape
```

```python
xtest.shape
```

### One Hot encode Y features

```python
from keras.utils import to_categorical
ytrain_cat=to_categorical(ytrain)
ytest_cat=to_categorical(ytest)
```

```python
ytrain_cat[0:5]
```

```python
ytest_cat[0:5]
```

### Create a neural network



```python
xtrain.shape[1]

```

```python
from keras.models import Sequential
from keras.layers import Dense,Dropout
from keras.callbacks import EarlyStopping
```

```python
model=Sequential()

# Layer 1
model.add(Dense(64,activation="relu",input_dim=xtrain.shape[1]))
model.add(Dropout(0.3))

# Layer 2
model.add(Dense(32,activation='relu'))
model.add(Dropout(0.2))

# Output layer
model.add(Dense(3,activation='softmax'))

```

```python
model.summary()
```

### Compile the model

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
early_stop=EarlyStopping(monitor='val_loss',patience=10)
```

```python
hist=model.fit(xtrain,ytrain_cat,validation_split=0.2,epochs=200,callbacks=[early_stop])
```

### Plot the learning curve

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['loss'],label='Trainig loss')
plt.plot(hist.history['val_loss'],label='Validation loss')
plt.xlabel('Epochs')
plt.ylabel('Categorical Cross Entropy')
plt.title('Learning Curve')
plt.show()

```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(16,8))
plt.plot(hist.history['accuracy'],label='Training accuracy')
plt.plot(hist.history['val_accuracy'],label='Validation accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Accuracy Curve')
plt.show()
```

### Evaluate model accuracy

```python
model.evaluate(xtrain,ytrain_cat)
```

```python
model.evaluate(xtest,ytest_cat)
```

### Predicting the results

```python
yprob_tr=model.predict(xtrain)
yprob_ts=model.predict(xtest)
```

```python
yprob_tr[0:5]
```

```python
yprob_ts[0:5]
```

### Convert probability to prediction

```python
import numpy as np 
ypred_tr=np.argmax(yprob_tr,axis=-1)
ypred_ts=np.argmax(yprob_ts,axis=-1)
```

```python
ypred_tr[0:5]
```

```python
ypred_ts[0:5]
```

### Plot the confusion matrix

```python
ytest
```

```python
from sklearn.metrics import confusion_matrix,ConfusionMatrixDisplay
cf=confusion_matrix(ytest,ypred_ts)
cfd=ConfusionMatrixDisplay(cf)
cfd.plot()
```

```python
from sklearn.metrics import classification_report
print(classification_report(ytest,ypred_ts))
```
